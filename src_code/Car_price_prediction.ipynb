```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

Adding data_set

```python
car_data = pd.read_csv('/content/CAR DETAILS FROM CAR DEKHO.csv')
```

```python
car_data.replace({'fuel':{'Diesel':0,'Petrol':1,'CNG':2,'LPG':3,'Electric':4},'seller_type':{'Individual':0, 'Dealer':1,'Trustmark Dealer':2}, 'transmission':{'Manual':0, 'Automatic':1}, 'owner':{'First Owner':0, 'Second Owner':1, 'Third Owner':2, 'Fourth & Above Owner':3, 'Test Drive Car':4}} , inplace=True)
```

```python
X = car_data.drop(columns=['name', 'selling_price'], axis=1)
Y = car_data['selling_price']
```

Implementing Linear regression model

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.1, random_state=2)
print(X.shape, x_train.shape, x_test.shape, y_train.shape)
model = LinearRegression()
model.fit(x_train, y_train)
```

implementing Lasso regression model

```python
model1 = Lasso()
model1.fit(x_train, y_train)
```

Evaluation of model accuracy

Linear regression accuracy

```python
linear= model.predict(x_train)
accuracy1 = metrics.r2_score(linear, y_train)
print(accuracy1)
```

Lasso accuracy

```python
lasso= model1.predict(x_train)
accuracy2 = metrics.r2_score(lasso, y_train)
print(accuracy2)
```
